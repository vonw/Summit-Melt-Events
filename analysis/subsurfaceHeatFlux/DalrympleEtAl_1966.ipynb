# Notes from Town's subsurface model

### Dalrymple's subsurface "material properties" from Town's subsurfaceHeatFluxAdvDiffTransBdry2D.m

```python
%pylab inline
import numpy as np
```

```python
z_Dal   = np.array([0.0, 0.5, 1.0, 1.5, 2.0, 2.5, 3.0, 3.5, 4.0, 4.5, 5.0, 5.5, 6.0, 6.5])   # m
rho_Dal = np.array([354, 373, 379, 377, 381, 393, 406, 413, 420, 422, 430, 446, 462, 478]) 	# kg/m^3

plot(rho_Dal, z_Dal)
gca().invert_yaxis()
xlabel('Snow Density (kg m-3)')
ylabel('Depth (m)')
title('Data from Dalrymple et al (1966)')
```

```python
cp_Dal  = np.array([0.147, 0.155, 0.157, 0.156, 0.158, 0.163, 0.168, 0.171, 0.174, 0.175, 0.178, 0.184, 0.190, 0.196])*4.187*100**3./rho_Dal  # converted from cal C^-1 cm^-3 to specific heat capacity (was volumetric heat capacity);

plot(cp_Dal, z_Dal)
gca().invert_yaxis()
xlabel('Snow Specific Heat Capacity (J kg-1 K-1)')
ylabel('Depth (m)')
title('Data from Dalrymple et al (1966)')
```

```python
K_Dal   = np.array([0.00282, 0.00317, 0.00364, 0.00416, 0.00466, 0.00512, 0.00557, 0.00593, 0.00616, 0.00622, 0.00622, 0.00622, 0.00622, 0.00622])/100**2 # cm^2/s to SI

plot(K_Dal, z_Dal)
gca().invert_yaxis()
xlabel('Thermal Diffusivity (m^2 s-1)')
ylabel('Depth (m)')
title('Data from Dalrymple et al (1966)')
```

```python
# Thermal conductivity, k, W m-1 K=1
k_Dal = K_Dal * rho_Dal * cp_Dal

plot(k_Dal, z_Dal)
gca().invert_yaxis()
xlabel('Thermal Conductivity (W m-1 K-1)')
ylabel('Depth (m)')
title('Data from Dalrymple et al (1966)')
```
